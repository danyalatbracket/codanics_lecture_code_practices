```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.metrics import mean_squared_error

# 1 Sample data
X = np.array([[500], [750], [1000], [1250], [1500]])  # Size of house (sq ft)
y = np.array([150, 200, 250, 300, 350])  # Price in thousands

# 2 Add another feature (e.g., number of rooms)
X = np.hstack([X, np.array([[2], [3], [4], [5], [6]])])

# 3 Linear Regression
lr = LinearRegression()
lr.fit(X, y)
y_pred_lr = lr.predict(X)

# 4 Ridge Regression (add regularization)
ridge = Ridge(alpha=1000)
ridge.fit(X, y)
y_pred_ridge = ridge.predict(X)

# 5 Print coefficients
print("Linear Coefficients:", lr.coef_)
print("Ridge Coefficients :", ridge.coef_)

# 6 Plot predictions
plt.figure(figsize=(10, 5))
plt.plot(y, label='Actual Prices', marker='o')
plt.plot(y_pred_lr, label='Linear Predicted', linestyle='--')
plt.plot(y_pred_ridge, label='Ridge Predicted', linestyle='--')
plt.legend()
plt.title("Ridge vs Linear Regression")
plt.xlabel("House Index")
plt.ylabel("Price (in 1000s)")
plt.grid(True)
plt.show()
```

# Noisy Data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, LinearRegression

# Generate linear data with noise
np.random.seed(0)
X = np.array([[500], [750], [1000], [1250], [1500]])
rooms = np.array([[2], [3], [4], [5], [6]])
noise = np.random.normal(0, 20, size=(5,))
y = np.array([150, 200, 250, 300, 350]) + noise

# Combine features
X = np.hstack([X, rooms])

# Fit Linear Regression
lr = LinearRegression()
lr.fit(X, y)
y_pred_lr = lr.predict(X)

# Fit Ridge Regression
ridge = Ridge(alpha=1000)
ridge.fit(X, y)
y_pred_ridge = ridge.predict(X)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(y, label='Actual Prices (Noisy)', marker='o')
plt.plot(y_pred_lr, label='Linear Predicted', linestyle='--')
plt.plot(y_pred_ridge, label='Ridge Predicted', linestyle='--')
plt.legend()
plt.title("Ridge vs Linear Regression (Noisy Data)")
plt.xlabel("House Index")
plt.ylabel("Price (in 1000s)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Seperated plots


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, LinearRegression

# Generate linear data with noise
np.random.seed(0)
X = np.array([[500], [750], [1000], [1250], [1500]])
rooms = np.array([[2], [3], [4], [5], [6]])
noise = np.random.normal(0, 20, size=(5,))
y = np.array([150, 200, 250, 300, 350]) + noise

# Combine features
X = np.hstack([X, rooms])

# Fit Linear Regression
lr = LinearRegression()
lr.fit(X, y)
y_pred_lr = lr.predict(X)

# Fit Ridge Regression
ridge = Ridge(alpha=1000)
ridge.fit(X, y)
y_pred_ridge = ridge.predict(X)

# Create figure with two subplots
plt.figure(figsize=(12, 5))

# First subplot - Linear Regression
plt.subplot(1, 2, 1)
plt.plot(y, label='Actual Prices', marker='o', color='blue')
plt.plot(y_pred_lr, label='Linear Regression', linestyle='--', marker='x', color='red')
plt.title("Linear Regression Predictions")
plt.xlabel("House Index")
plt.ylabel("Price (in 1000s)")
plt.legend()
plt.grid(True)

# Second subplot - Ridge Regression
plt.subplot(1, 2, 2)
plt.plot(y, label='Actual Prices', marker='o', color='blue')
plt.plot(y_pred_ridge, label='Ridge Regression', linestyle='--', marker='x', color='green')
plt.title("Ridge Regression Predictions")
plt.xlabel("House Index")
plt.ylabel("Price (in 1000s)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
