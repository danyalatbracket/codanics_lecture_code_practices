```python
# Loops and Iteration
# Create a program that:

# Asks the user for a number, n.
# Uses a for loop to print the multiplication table of n (e.g., for 5: 5, 10, 15, …, 50)

number = int(input("Please Enter any number"))

print(f"Table of {number} is:")
for i in range(1, 11):
    print(f"{number} * {i} = ", number * i)
```

```python
# 4. Lists
# Write a program to:

# Create a list of 5 integers entered by the user.
# Print the sum of all numbers in the list.
# Find and print the smallest and largest numbers in the list.

num_List = []
for i in range(5):
    number = int(input(f"Please Enter Number {i+1}"))
    num_List.append(number)

print("List of numbers enterd by the user is")
print(num_List)
counter = 0
sum = 0
min = 0
max = 0

for i in num_List:

    if(counter == 0):
        min = i
        max = i

    if i < min:
        min = i
    elif i > max:
        max = i

    sum += i
    counter += 1
    
print("Sum of all integers in the list is = ", sum)
print(f"Max number is {max} and Min number is {min}")
```

```python
# Functions
# Write a function called is_prime() that:

# Takes an integer as input.
# Returns True if the number is prime, otherwise returns False.

def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, number):
        if number % i == 0:
            return False

    return True


print(is_prime(17))
print(is_prime(27))
print(is_prime(29))
print(is_prime(35))
print(is_prime(39))


# improved code by gpt

# import math

# def is_prime(number):
#     if number <= 1:
#         return False
#     for i in range(2, int(math.sqrt(number)) + 1):  # Only go up to the square root of number
#         if number % i == 0:
#             return False
#     return True

# # Test cases
# print(is_prime(17))  # True
# print(is_prime(27))  # False
# print(is_prime(29))  # True
# print(is_prime(35))  # False
# print(is_prime(39))  # False
```

```python
# DataFrame Operations (with Pandas)
# Write a program to:

# Create a DataFrame with the following data:

# Name	Age	Marks
# Alice	23	89
# Bob	25	76
# Charlie	22	92

# Perform the following tasks:
# Display the first two rows of the DataFrame.
# Add a new column, Grade, where:
# Marks ≥ 90: A
# Marks between 80 and 89: B
# Marks below 80: C

import pandas as pd

Names = ["Alice", "Bob", "Charlie"]
Ages = [23, 25, 22]
Marks = [89, 76, 92]

df = pd.DataFrame()
df["Names"] = Names
df["Age"] = Ages
df["Marks"] = Marks

first_two_rows = df.head(2)
print(first_two_rows)

print('=========================================')
# Add a new column, Grade
def assign_grade(marks):
	if marks >= 90:
		return 'A'
	elif marks >= 80:
		return 'B'
	else:
		return 'C'

df['Grade'] = df['Marks'].apply(assign_grade)
print(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a list of numbers
data = [10, 20, 15, 22, 100, 35, 25]
data = sorted(data)
# Plot a boxplot using Matplotlib
plt.figure(figsize=(10, 7))
plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

# Identify outliers based on the boxplot
data_np = np.array(data)
Q1 = np.percentile(data_np, 25)
Q3 = np.percentile(data_np, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data_np[(data_np < lower_bound) | (data_np > upper_bound)]
# this above expression transaltes to below expression and returns the ouliers array in the data
# outliers = data_np[np.array([False, False, False, False, False, False, True])]

print(f"Outliers: {outliers}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate a random dataset with 3 clusters and 2 features
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Step 2: Perform K-Means clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Get the cluster centers and the labels for each point
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Step 3: Visualize the clusters
plt.figure(figsize=(10, 7))


# Plot all the data points with colors based on the cluster label
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)

# Plot the cluster centers in red with a larger 'X' marker
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, marker='X', label='Centroids')

# Add title and labels
plt.title("K-Means Clustering (3 Clusters)", fontsize=16)
plt.xlabel("Feature X")
plt.ylabel("Feature Y")
plt.legend()
plt.show()
```
